## Scatter Plot
  - X-axis: 'total_bill' column from the `tips` dataset.
  - Y-axis: 'tip' column from the `tips` dataset.
  - Hue: Differentiates data points based on the 'day' column.
  - Size: Adjusts marker sizes based on the 'size' column.
  - Palette: Utilizes the "YlGnBu" color palette.

```python
# Create scatter plot
sns.scatterplot(
    x="total_bill", 
    y="tip", 
    data=tips, 
    hue="day", 
    size="size", 
    palette="YlGnBu"
)
```

```python
sns.histplot(tips['tip'], kde=True,bins=15)
```

```python
sns.displot(tips['tip'], kde=True,bins=15)
```

```python
sns.barplot(x='sex', y='tip', data=tips, palette='YlGnBu')
```

## Boxplot
